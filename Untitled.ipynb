```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
plt.figure(figsize=(12, 8))
plot_tree(dt, filled=True, feature_names = X.columns)
```

```python
print(export_text(dt, feature_names=X.columns, show_weights=True))
```

```python
dt_pred = dt.predict(X_test)
```

```python
dt_acc = accuracy_score(y_test, dt_pred)
```

Bagging, random forest and boosting

Bagging

```python
dt_bag = RandomForestClassifier(n_estimators=500, max_features=None, random_state=1)
dt_bag.fit(X_train, y_train)
dt_bag_pred = dt_bag.predict(X_test)
dt_bag_acc = accuracy_score(y_test, dt_bag_pred)
confusion_matrix(y_test, dt_bag_pred)
```

```python
feature_importances_bag = dt_bag.feature_importances_
# This is a numpy array

# Sort feature importances in descending order
sort_index_bag = np.argsort(feature_importances_bag)
# sort_index = sort_index[::-1]
# argsort() returns the index position that the element would have 
# if the array were sorted in ascending order.
# We can use it to select feature importance and 

# Plot horizontal feature importances
plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.barh(range(X.shape[1]), feature_importances_bag[sort_index_bag], align="center")
plt.yticks(range(X.shape[1]), np.array(X.columns)[sort_index_bag])
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.tight_layout()
```

Random Forest

```python
dt_rf = RandomForestClassifier(n_estimators=500, max_features = 4, random_state=0)
# max_features is set to sqrt(p), about 4.

dt_rf.fit(X_train, y_train)
dt_rf_pred = dt_rf.predict(X_test)
dt_rf_acc = accuracy_score(y_test, dt_rf_pred)
confusion_matrix(y_test, dt_rf_pred)
```

We cannot get an interpretable tree from ensemble methods, but we can figure out the importance of each variable in classification.

```python
feature_importances = dt_rf.feature_importances_
# This is a numpy array

# Sort feature importances in descending order
sort_index = np.argsort(feature_importances)
# sort_index = sort_index[::-1]
# argsort() returns the index position that the element would have 
# if the array were sorted in ascending order.
# We can use it to select feature importance and 

# Plot horizontal feature importances
plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.barh(range(X.shape[1]), feature_importances[sort_index], align="center")
plt.yticks(range(X.shape[1]), np.array(X.columns)[sort_index])
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.tight_layout()
```

Boosting

```python
dt_gb = GradientBoostingClassifier(n_estimators=100, learning_rate =0.001,
max_depth=3, random_state=0)
# B = 100, lambda = 0.001, d = 3
# Too large B may lead to overfitting
dt_gb.fit(X_train, y_train)
dt_gb_pred = dt_gb.predict(X_test)
dt_gb_acc = accuracy_score(y_test, dt_gb_pred)
confusion_matrix(y_test, dt_gb_pred)
```

```python
feature_importances_gb = dt_gb.feature_importances_
# This is a numpy array

# Sort feature importances in descending order
sort_index_gb = np.argsort(feature_importances_gb)
# sort_index = sort_index[::-1]
# argsort() returns the index position that the element would have 
# if the array were sorted in ascending order.
# We can use it to select feature importance and 

# Plot horizontal feature importances
plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.barh(range(X.shape[1]), feature_importances_gb[sort_index_gb], align="center")
plt.yticks(range(X.shape[1]), np.array(X.columns)[sort_index_gb])
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.tight_layout()
```

```python
print('-----------Decision Tree-----------')
print(dt_acc)
print('\n')
print('-----------Bagging-----------')
print(dt_bag_acc)
print('\n')
print('-----------Random Forest-----------')
print(dt_rf_acc)
print('\n')
print('-----------Boosting-----------')
print(dt_gb_acc)
```
