```python
import pandas as pd
import statsmodels.api as sm
```

```python
toyota = pd.read_csv('ToyotaCorolla.csv')
```

```python
toyota.head()
```

```python
toyota.info()
```

Since fuel_type is object, we cannot consider as dummy variables; We need to convert to int and then use it as dummy variables

```python
toyota.describe().T
```

```python
y = toyota['Price']
X = toyota[['Age','KM','HP','MetColor','Automatic']]

X_with_intercept = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X_with_intercept)
results = model.fit()
```

```python
results.summary()
```

For the two dummy variables:
Metcolor : Comparing to cars with no metallic color, cars with it sells 44 dollars more, holding evrything else constant. But its not significant.

Automatic: Comparing to cars with no auto transmission, cars with it sells for 748 dollars more, holding everything else constant

```python
#Change the variable data type from integer to categorical
toyota['Doors'].unique()
toyota['Doors'] = toyota['Doors'].astype('category')
```

```python
toyota.info()
```

```python
#Generate dummy variables
X_car_1 = toyota[['Age','KM','HP','MetColor','Automatic','Doors','Fuel_Type']]
X_car_1 = pd.get_dummies(X_car_1, columns = ['Doors','Fuel_Type'], drop_first=True)
```

```python
X_car_1.head()
```

```python
X_car_1.info()
```

```python
X_with_intercept_car1 = sm.add_constant(X_car_1)

# Fit the linear regression model
model_car = sm.OLS(y, X_with_intercept_car1)
results_car = model_car.fit()
```
